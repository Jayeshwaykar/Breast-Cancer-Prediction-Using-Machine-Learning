```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
breast_cancer_data = pd.read_csv("/content/drive/MyDrive/Breast Cancer Classification/Breast Cancer Data.csv")
```

```python
breast_cancer_data.drop(['Unnamed: 32'],axis=1)
```

```python
# Statistical measures about the data
breast_cancer_data.describe()
```

```python
breast_cancer_data['label'] = breast_cancer_data['label'].replace(['M'], '0')
```

```python
breast_cancer_data['label'] = breast_cancer_data['label'].replace(['B'], '1')
```

```python
breast_cancer_data
```

```python
breast_cancer_data['label'].value_counts()
```

```python
breast_cancer_data.groupby('label').mean()
```

Separating the features and target

```python
X = breast_cancer_data.drop(['label','id','Unnamed: 32','diagnosis'],axis=1)
Y = breast_cancer_data['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic regression

```python
model = LogisticRegression()
```

```python
# training the logistic regression model
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(Y_train, X_train_prediction)
```

```python
print("Accuracy on training data =",training_data_accuracy)
```

```python
# accuracy on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy= accuracy_score(Y_test, X_test_prediction)
```

```python
print("Accuracy on test data =",test_data_accuracy)
```
